```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
data = pd.read_csv('Consumo_cerveja.csv', sep = ",")
```

```python
data.head()
```

```python
#Verifica se há dados faltantes no data set
data.isna() .any()
```

```python
#soma a quantidade de dados faltantes no data set
data.isna() .sum()
```

```python
#Elimina as celulas vazias do dataset
data = data.dropna()
```

```python
# Exibe as estatisticas descritivas do Dataset

data.describe()
```

```python
#Exibe a o tipo de dado por coluna

data.info()
```

```python
#renomear colunas e salvar na variável

data = data.rename(columns = {'Temperatura Media (C)': 'temp_media', 'Temperatura Minima (C)': 'temp_min', 'Temperatura Maxima (C)': 'temp_max', 'Final de Semana': 'fds', 'Consumo de cerveja (litros)': 'consumo' })
```

```python
data = data.astype({'fds': 'float'})
```

```python
data['Data'] = pd.to_datetime(data['Data'])
```

```python
data.head()
```

```python
data['temp_media'] = data['temp_media'].astype(float)
```

```python
#Converte todos os campos em float

date = data.replace('[^\d.]', '', regex=True).astype(float)
```

```python
date.info()
```

```python
date['Data'] = pd.to_datetime(date['Data'])
```
