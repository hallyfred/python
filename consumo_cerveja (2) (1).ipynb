```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
data = pd.read_csv('Consumo_cerveja.csv', sep = ",")
```

```python
#Verifica se há dados faltantes no data set
data.isna() .any()
```

```python
#soma a quantidade de dados faltantes no data set
data.isna() .sum()
```

```python
#Elimina as celulas vazias do dataset
data = data.dropna()
```

```python
# Exibe as estatisticas descritivas do Dataset

data.describe()
```

```python
#Exibe a o tipo de dado por coluna

data.info()
```

```python
#renomear colunas e salvar na variável

data = data.rename(columns = {'Temperatura Media (C)': 'temp_media', 'Temperatura Minima (C)': 'temp_min', 'Temperatura Maxima (C)': 'temp_max', 'Final de Semana': 'fds', 'Consumo de cerveja (litros)': 'consumo', 'Precipitacao (mm)': 'chuva' })
```

```python
#Converte todos os campos em float

date = data.replace('[^\d.]', '', regex=True).astype(float)
```

```python
date['Data'] = pd.to_datetime(date['Data'])
```

```python
date.describe() .round(2)
```

```python
date.corr() .round(4)
```

```python
# chamando a função subplots; cria uma figura e permite que você 
# insira gráficos dentro dessa figura
fig, ax = plt.subplots(figsize=(20,6))

# adicionando um título ao nosso gráfico
ax.set_title('Consumo de Cerveja ao longo de 365 dias', fontsize=20)

# identificando os eixos
ax.set_ylabel('Consumo (litros)', fontsize=14)
ax.set_xlabel('Tempo (dias)', fontsize=14)

# plotando o nosso gráfico; o parâmetro fontsize vai definir o tamanho 
# da fonte da gradação dos eixos
ax = date['consumo'].plot(fontsize=14)
```

```python
# plotando um histograma para a variável 'consumo'
ax = sns.distplot(date['consumo'])
# fazendo formatações
ax.figure.set_size_inches(12,6)
# adicionando um título ao nosso gráfico
ax.set_title('Distribuição de Frequências de Consumo de Cerveja', fontsize=20)
# identificando os eixos
ax.set_xlabel('consumo', fontsize=14)
ax.set_ylabel('Frequência', fontsize=14)
```
